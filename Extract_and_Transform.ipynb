```python
import pandas as pd
import json
df=pd.read_csv('/content/restaurants_small.csv')
df.head()
```

```python
# df1=pd.DataFrame(json.dumps(df.iloc[0]['items']))
df1 = pd.DataFrame.from_dict(json.loads(df.iloc[0]['items']), orient="index")
df1=df1.reset_index()
df1['Resturant_id']=df.iloc[0]['id']
# "{x} is even" if x %2==0 else f"{x} is odd
df1['item_details']=df1.apply(lambda x: x['index'].split('[')[1][:-1] if len(x['index'].split('[')) > 1 else "", axis=1)
df1['index']=df1.apply(lambda x: x['index'].split('[')[0] if len(x['index'].split('[')) > 1 else x['index'], axis=1)
# a=json.loads(df.iloc[0]['full_details'])
# if 'currency' in a:
#   df1['Currency']=a['currency']
for i in range(1,df.shape[0]):
  df2 = pd.DataFrame.from_dict(json.loads(df.iloc[i]['items']), orient="index")
  df2=df2.reset_index()
  df2['Resturant_id']=df.iloc[i]['id']
  df2['item_details']=df2.apply(lambda x: x['index'].split('[')[1][:-1] if len(x['index'].split('[')) > 1 else "", axis=1)
  df2['index']=df2.apply(lambda x: x['index'].split('[')[0] if len(x['index'].split('[')) > 1 else x['index'], axis=1)
  # try:
  #   a=json.loads(df.iloc[i]['full_details'])
  #   if 'currency' in a:
  #     df2['Currency']=a['currency']
  # except:
  #   pass
  df1=pd.concat([df1, df2])
  # df1.append(df2,ignore_index = True)
```

```python
import pandas as pd
import json
df=pd.read_csv('/content/restaurants_small.csv')
df['Latitude']= df.apply(lambda x: x['lat_long'].split(',')[0] if len(x['lat_long'].split(',')) > 1 else "", axis=1)
df['Longitude']= df.apply(lambda x: x['lat_long'].split(',')[1] if len(x['lat_long'].split(',')) > 1 else "", axis=1)
df['Rating']=0.0
df['No_of_votes']=0
for i in range(0,df.shape[0]):
  try:
    a=json.loads(df.iloc[i]['full_details'])
    if 'user_rating' in a:
      df.iat[i,8]=a['user_rating']['aggregate_rating']
      df.iat[i,9]=a['user_rating']['votes']
  except:
    pass
df['True_rating']= 0.45*df['Rating'].astype(float) + 0.55*df['No_of_votes'].astype(float)
df['True_rating'] = (df['True_rating'] - df['True_rating'].min()) / (df['True_rating'].max() - df['True_rating'].min())
df.head()
```

```python
df=df.drop(['items','lat_long','full_details'],axis=1)
df.head(20)
```

```python
df.to_csv('resturant_details.csv')
```
